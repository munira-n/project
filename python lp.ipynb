```python
#I decided to make graphs based on dataset provided by kaggle.com. I chose world universeties ranking in four different years 
#2012-2015

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.axes_grid1 import Divider, Size

df = pd.read_csv ('dataset.csv')
```

```python
print(df)
```

```python
dk=df[['world_rank','institution', 'year']]
dk.head()
```

```python
wr_sort = dk.sort_values('world_rank')
wr_sort = wr_sort.loc[0:309,['world_rank','institution','year']]
wr_sort
```

```python
ysort = wr_sort.sort_values('year')
ysort.loc[:, ['world_rank','institution','year']]
ysort
```

```python
inst_count=wr_sort['institution'].value_counts()
inst_count.iloc[0:10]
```

```python
country = df[['country']]
country = country.sort_values('country')
count = country.value_counts()
count = count.iloc[:10]
count
```

```python
my_plot = count.plot(kind='bar', figsize=(15, 10))

font1 = {'family':'serif','color':'aquamarine','size':20}
font2 = {'family':'serif','color':'magenta','size':15}
my_plot.set_xlabel("Countries", fontdict = font2)
my_plot.set_ylabel("Frequency occurence", fontdict = font2)
plt.title('10 most occured countries 2012-2015', fontdict = font1, loc = 'center')
plt.tight_layout()
```
